```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file
file_name =(r'Student\speech-emotion-recognition-ravdess-data\Actor_01\03-01-01-01-01-01-01.wav') # Replace with your file
y, sr = librosa.load(file_name, sr=None)

# Extract MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Create a heatmap for the MFCCs
plt.figure(figsize=(12, 6))
librosa.display.specshow(mfccs, sr=sr, x_axis='time', cmap='coolwarm')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCCs')
plt.tight_layout()
plt.xlabel('Time (frames)')
plt.ylabel('MFCC Coefficients')
plt.show()
```
